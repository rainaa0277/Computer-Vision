# ROTATION

```python
def rotate(image, angle, center=None, scale=1.0):
    (h,w)=image.shape[:2]
    
    if center == None:
        center= (w//2, h//2)
        
        #create matrix for rotation
        
        M = cv2.getRotationMatrix2D(center, angle, scale)
        rotated = cv2.warpAffine(image, M, (w,h))
        
        return rotated
```

```python
rot = rotate(jwhd, 120)
func(rot)
```

# Aspect Ratio

```python
def resize(image, width= None, height=None):
    dim = None
    (h,w) = image.shape[:2]
    
    if width is None and height is None:
        return image
    #calculate new dimension based on the height
    elif width is None:
        r = height/float(h)
        dim = (int(w*r), height)
        
    #calculate new dimension based on the width
    elif height is None:
        r = width/float(w)
        dim = (int(h*r), width)
    
    res = cv2.resize(image, dim)
    
    return res
        
```
